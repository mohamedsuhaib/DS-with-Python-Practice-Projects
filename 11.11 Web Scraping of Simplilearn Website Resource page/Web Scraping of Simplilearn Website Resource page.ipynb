## Web Scraping of Simplilearn Website Resource page

### DESCRIPTION

__What to:__

__Scrape the Simplilearn website resource page and perform the following tasks:__

+ View and print the Simplilearn web page content in a well formatted format
+ View the head and title
+ Print all the href links present in the Simplilearn web page
+ Search and print the resource headers of the Simplilearn web page
+ Search resource topics
+ View the article names and navigate through them

Simplilearn website URL: http://www.simplilearn.com/resources

#### 1: View and print the Simplilearn webpage content in a good format

```python
#Import the required libraries
from bs4 import BeautifulSoup
#import urllib2 / instead of this executing below command
from urllib.request import urlopen
```

```python
#Provide the url for web scraping
url = 'https://www.simplilearn.com/resources'
```

```python
#Scrape the web page document
#webpage - urllib2.urlopen(url # change this step to below one
webpage = urlopen(url)
```

```python
#Create the soup object and parse the webpage using an html parser
sl_soup = BeautifulSoup(webpage,"html.parser")
```

```python
#Close the webpage object
webpage.close()
```

```python
#View the soup contents
print (sl_soup.contents)
```

```python
#Use the prettify() method to view the well-formatted html output
print (sl_soup.prettify())
```

#### 2: View the title

```python
#View the title of the scraped document
print (sl_soup.title)
```

#### 3: Print all the href links present in the Simplilearn webpage

```python
#Find all href links
for href in sl_soup.findAll("a", href=True):
    print (href["href"])
```

#### 4: Search and print the resource headers of the Simplilearn webpage 

```python
#View all the headers name
#the div information is obtained from the web page (view page source)
for article in sl_soup.findAll("h2"):
    print(article.string)
```

#### 5: Search resource topics

```python
#View all the resource topics
#the class information is obtained from the web page (view page source)
for article_topic in sl_soup.findAll("h4"):
    print(article_topic.string)
```

#### 6: View the article names and navigate through them

```python
#Create article object for finding the list (Hint: the class information is obtained from the web page (view page source))
url2="https://www.simplilearn.com/tutorials/project-management-tutorial/what-is-agile-project-management?source-frs_home"
webpage2 = urlopen(url2)
sl_article = BeautifulSoup(webpage2,"html.parser")
#Close the webpage2 object
webpage2.close()
test = sl_article.find(class_="desig_author empty-text")
```

```python
#View the type of the article object
type(test)
```

```python
#Find all the article names
test.findAll('p')
```

```python
#Find first article name
first_next_p = test.p
first_next_p
```

```python
#Find next sibling (next article name)
find_next_sibling = first_next_p.next_sibling.next_sibling
find_next_sibling
```

```python
#Find previous sibling (previous article name)
find_next_sibling.previous_sibling.previous_sibling
```
